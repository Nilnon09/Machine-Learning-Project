```python
import ast
ast.literal_eval('[{"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 28, "name": "Action"}]')
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
     l.append(i['name']) 
    return l   
```

```python
def convert1(obj):
    l=[]
    count=0
    for i in ast.literal_eval(obj):
        if count<3:
            l.append(i['name']) 
            count+=1
        else:
         break    
    return l   
```

```python
mov['cast']=mov['cast'].apply(convert1)
```

```python
def fetch_director(obj):
    k=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            k.append(i['name']) 
            break    
    return k   
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
ps.stem('loving')
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
df['new_tags'].apply(stem)
```

```python
df['new1tags']=df['new_tags'].apply(stem)
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(df['new_tags']).toarray()
```

```python
def recomend(movie):
    movie_index=df[df['title']==movie].index[0]
    distance=similar[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(df.iloc[i[0]].title)
        
   
    
    
    
```
